## Visualization & Explanatory Analysis

### 1 Bagaimana Tren Curah Hujan di Nongzhanguan dalam Periode 2013 hingga 2017?


```python
Nongzhanguan_df.set_index("datetime", inplace=True)
```

```python
Nongzhanguan_df.sample(5)
```

```python
rain_years = Nongzhanguan_df.resample("YE").mean(numeric_only=True)

# Plot curah hujan
plt.figure(figsize=(10, 5))
plt.plot(rain_years.index, rain_years["RAIN"], marker="o", linestyle="-", color="g")

plt.xlabel("Tahun")
plt.ylabel("Curah Hujan Rata-rata (mm)")
plt.title("Curah Hujan Tahunan di Nongzhanguan")
plt.grid()
plt.show()
```

### Pertanyaan 2: Bagaimana Tekanan pada Atmosfer pada wilayah Nongzhanguan dalam periode tahun 2013 hingga 2017?


```python
temp_years = Nongzhanguan_df.resample("YE").mean(numeric_only=True)

plt.figure(figsize=(10, 5))
plt.plot(temp_years.index, temp_years["PRES"], marker="o", linestyle="-", color="b")

plt.xlabel("Tahun")
plt.ylabel("Tekanan Rata-rata Atmosfer(Hpa)")
plt.title("Tekanan Rata-Rata Atmosfer Tahunan di Nongzhanguan")
plt.grid()
plt.show()

plt.show()
```

### Pertanyaan 3: Bagaimana Temperatur wilayah Nongzhanguan dalam periode tahun 2013 hingga 2017?

```python
temp_years = Nongzhanguan_df.resample("YE").mean(numeric_only=True)

plt.figure(figsize=(12, 5))
plt.plot(temp_years.index, temp_years["TEMP"], marker="o", linestyle="-", color="b")

plt.xlabel("Tahun")
plt.ylabel("Temperatur Rata-rata (°C)")
plt.title("Temperatur Tahunan di Nongzhanguan")
plt.grid()
plt.show()

plt.show()
```

**Insight:**
- curah hujan di stasiun nongzhanguan mengalami peningkatan pada 2017 terlihat pada 2013 - 2015 m3ngalami penuruan curah hujan. di awal tahun 2015 - 2017 mengalami peningkatan curah hujan yang cukup extrem. Peningkatan ini bisa mengindikasikan perubahan pola cuaca, seperti musim hujan yang lebih panjang atau intensitas hujan yang lebih tinggi.
- tekanan udara pada stasiun nongzhanguan cenderung stabil di tahun 2014 - 2017 dan meningkat di awal tahun 2017 sampai awal tahun 2018
-temperatur udara pada stasiun nongzhanguan di 2014 - 2017 terus menurun dan mengalami temperatur relatif stabil pada awal tahun 2015 - 2017. menurun extrem diawal tahun 2017 - 2018
- Penurunan Temperatur cuaca, tekanan udara dan curah hujan yang terus meningkat sepanjang tahun 2017 bisa menjadi indikasi bahwa terjadi perubahan pola pada atmosfer dan musim dingin.
